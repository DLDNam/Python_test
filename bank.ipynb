```python
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
my_df = pd.read_csv("bank.csv")
my_df.head(30)
```

```python
df_marital = my_df['marital'].value_counts()
```

```python

sizes = []
labels = []
for x,y in df_marital.items():
    labels.append(x)
    sizes.append(y)
colors = ['yellowgreen', 'lightskyblue', 'lightcoral']


plt.pie(sizes, labels=labels, colors=colors, autopct='%.2f%%')
plt.axis('equal')

plt.show()
```

```python
my_duration = my_df.groupby('job').sum()['duration'].sort_values(ascending= False)
my_duration
```

```python
job = []
duration = []
for x,y in my_duration.items():
    job.append(x)
    duration.append(y)
```

```python
plt.bar(height=duration, x=job )
plt.xticks(rotation = 90)
plt.show()
```
