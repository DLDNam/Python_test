```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
my_df = pd.read_csv("sales2019_7.csv")
```

```python
my_df = my_df[my_df['Product'] !='Product']
my_df = my_df.dropna(how = 'all')
```

```python
set(my_df['Product'])
```

```python
my_df.head(30)
```

```python
my_df['Quantity Ordered'] = pd.to_numeric(my_df['Quantity Ordered'],downcast='integer')
my_df['Price Each'] = pd.to_numeric(my_df['Price Each'], downcast='float')
my_df['Sales'] = my_df['Price Each'] * my_df['Quantity Ordered']
my_df.head(5)
```

```python
sale = my_df.pop('Sales')
my_df.insert(4,'Sales',sale)
```

```python
Prd = my_df.groupby('Product').sum()['Sales'].sort_values(ascending= True)
```

```python
my_product = [x for x,y in Prd.items() ]
my_Sales = [y for x,y in Prd.items()]
```

```python
plt.bar( x = my_product, height = my_Sales, color = 'red')
plt.title("bass")
plt.xticks(rotation = 90)
plt.grid()
plt.show()
```

```python
def split_city(x):
    x= str(x)
    city = x.split(',')
    return city[1]
my_df['City'] = my_df['Purchase Address'].apply(lambda x:split_city(x))
```

```python
rvn_city = my_df.groupby('City').sum()['Sales']
city = []
revenue_ct = []
for x,y in rvn_city.items():
    city.append(x)
    revenue_ct.append(y)
print(city)
print(revenue_ct)
```

```python
def split_time(time):
    time = str(time)
    row_time = time.split('/')
    row_time = row_time[2].split(' ')
    row_time = row_time[1].split(':')
    return row_time[0]
my_df['Hours'] = my_df['Order Date'].apply(lambda x: split_time(x))




```
